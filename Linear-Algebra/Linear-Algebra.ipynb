# Vector Norms

The length of the vector is referred to as the vector norm or the vector's magnitude.

one norm

 
 
 
 P-norm
 
 
 
 
 infinity norm
 
 

```python
#L1 norm

import numpy as np
from numpy.linalg import norm
a=np.array([1,2,3])
l1_norm=norm(a,1)
print(l1_norm)
```

```python
#L2 norm

b=np.array([1,2,3,4,5,6,7,8])
l1_norm=norm(b,1)
print(l1_norm)
```

```python
l2_norm=norm(b,2)
print(l2_norm)
```

```python
c=np.array([1,2])
l2_norm=norm(c,2)
print(l2_norm)
```

```python
from math import inf
l_norm_infinity=norm(b,inf)
```

```python
print(l_norm_infinity)
```
